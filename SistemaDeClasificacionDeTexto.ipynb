```python
from bs4 import BeautifulSoup
import urllib.request

# Configura una solicitud con un encabezado User-Agent para imitar un navegador
url = 'https://librefinanciero.com'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36'}
request = urllib.request.Request(url, headers=headers)

# Abre la URL y parsea el contenido
response = urllib.request.urlopen(request)
html = response.read()
soup = BeautifulSoup(html, "html5lib")
text = soup.get_text(strip=True)
print(text)
```

**Tokenizar Texto**

```python
# Creo los Tokens por palabras
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text,'spanish')
tokens = [word.lower() for word in tokens if word.isalpha()] # remueve los signos de puntuacion
print(tokens)
```

```python
# Verifico frecuencia de palabras
freq = nltk.FreqDist(tokens)
for key,val in freq.items():
    print(str(key) + ':' + str(val))
```

```python
# Visualizo Tokens
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
freq.plot(30,cumulative=False)
```

**Stop Word**

```python
from nltk.corpus import stopwords
clean_tokens = tokens[:]

for token in tokens:

    if token in stopwords.words('spanish'):
        clean_tokens.remove(token)
print(clean_tokens)
```

```python
# Verfico nuevamente frecuencia de palabras
freq_clean = nltk.FreqDist(clean_tokens)
for key,val in freq_clean.items():
    print(str(key) + ':' + str(val))
```

```python
# Visualizo Tokens nuevamente
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
freq_clean.plot(30,cumulative=False)
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.corpus import wordnet
synonyms = []
for syn in wordnet.synsets('investment'):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())
print(synonyms)
```

**Obtener Sin√≥nimos**

```python
# Reemplazo tokens sin√≥nimos
for ind,si in enumerate(synonyms):
  clean_tokens_sin = [word.replace(synonyms[ind],'investment') for word in clean_tokens]
```

```python
# Reemplazar Tokens sin√≥nimos manual
sinonimos = ['libertad', 'independencia']

for ind,si in enumerate(sinonimos):
  clean_tokens_sin = [word.replace(sinonimos[ind],'libertad') for word in clean_tokens]
```

```python
# Recalculo frecuencia de palabras con sin√≥nimos agregados
#freq_clean_sin
fleq_clean_sin = nltk.FreqDist(clean_tokens_sin)
for key,val in fleq_clean_sin.items():
    print(str(key) + ':' + str(val))
```

```python
# Visualizaci√≥n Tokens
fleq_clean_sin.plot(30, cumulative = False)
```

**Obtener Sin√≥nimos**

```python
from nltk.corpus import wordnet
anonyms = []
for syn in wordnet.synsets('good'):
    for l in syn.lemmas():
      if l.antonyms():
        anonyms.append(l.antonyms()[0].name())
print(anonyms)
```
